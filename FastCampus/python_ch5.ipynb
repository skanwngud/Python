### 재귀함수
자기 스스로를 호출하는 함수

```python
def func(count):
    if count > 0:  # count 크기 체크
        print(count, '현재')  # 조건에 따른 값 출력
        func(count-1)  # count 의 값이 변화 된 채로 다시 func 호출
    print('결과', count)  # 스택처럼 쌓여있던 호출 된 func 의 값들이 차례대로 다시 실행 됨
```

```python
func(10)
```

재귀함수의 호출 단계가 깊어질수록 메모리를 추가적으로 사용하므로 종료조건을 분명히 해야한다.

```python
def no_idea():
    print("나는 아무 생각이 없다.")
    print("왜냐하면 아무 생각이 없기 때문이다.")
    no_idea()
```

```python
no_idea()
```

### 중첩함수
함수 안에 정의 된 함수

```python
def func1(a):
    def func2():
        nonlocal a  # 중첩함수에서 쓰이는 global a 와 같은 느낌, 한 단계 위에 있는 함수의 매개변수를 가져온다.
        a += 1
        return a
    return func2()
func1(2)
```

### lambda 함수
이름이 없는 한 줄 짜리 함수
lambda 인수 나열: 반환할 내용이며 인수가 없으며 생략 가능

e.g.) lambda x,y: x+y

```python
# 함수의 인수로 함수를 받을 때 람다 함수를 사용 할 수도 있다.
# 다른 함수와 연계할 때에도 활용 가능.
def run(func, x):
    print(func(x))
    
run(lambda x:x+2, 2)
```

### zip, map, filter

zip(리스트1, 리스트2): 두 쌍의 리스트의 원소를 하나의 튜플 형태로 묶어줌

```python
list1 = [1, 2, 3, 4]
list2 = [5, 6, 7, 8]

list(zip(list1, list2))
```

map(함수, collection): collection 값들을 함수의 인자로 넣어 연산한 값들을 반환

```python
list(map(lambda x:x+1, [1, 2, 3, 4]))
```

filter(함수, collection): collection 값들을 함수의 인자로 넣어 연산한 값들이 True 인 경우만 반환

```python
list(filter(lambda x:x>2, [4, 3, 2, 1]))
```
