```python
# subplot, subplots
# 2개 이상의 차트 그리기
```

```python
import matplotlib.pyplot as plt
```

```python
x1, y1 = [1, 2, 3], [1, 2, 3]
x2, y2 = [1, 2, 3], [1, 100, 200]
```

```python
# subplot (행의 수, 열의 수, 해당 그래프가 그려질 위치)
```

```python
plt.subplot(1, 2, 1) # 행 하나, 열 2개 중에 첫 번째
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(1, 2, 2) # 행 하나, 열 2개 중에 두번째
plt.plot(x2, y2)
plt.title('data2')
```

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('data2')
```

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('data2')

plt.tight_layout() # layout 최적화
```

```python
x1, y1 = [1, 2, 3], [1, 2, 3]
x2, y2 = [1, 2, 3], [1, 50, 200]
```

```python
# y축을 2개 가진 차트

fig, axe1 = plt.subplots()

axe2 = axe1.twinx() # 복사
axe1.plot(x1, y1, 'r', label='data1')
axe2.plot(x2, y2, 'b', label='data2')

axe1.set_xlabel('x', fontsize=15)
axe1.set_ylabel('y1', fontsize=15)
axe2.set_ylabel('y2', fontsize=15)
```

```python
fig, axe1 = plt.subplots()

axe2 = axe1.twinx() # 복사
chart1 = axe1.plot(x1, y1, 'r')
chart2 = axe2.plot(x2, y2, 'b')

axe1.set_xlabel('x', fontsize=15)
axe1.set_ylabel('y1', fontsize=15)
axe2.set_ylabel('y2', fontsize=15)

chart = chart1 + chart2
axe1.legend(chart, ['data1', 'data2'])
```
