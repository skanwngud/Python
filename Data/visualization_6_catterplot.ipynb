```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data=df, x='sepal_width', y='sepal_length')
```

```python
sns.scatterplot(data=df, x='sepal_width', y='sepal_length', hue='species')
```

```python
# column 분포
df['sepal_length'].unique()
```

```python
def function1(x):
    if x<3:
        return 's'
    elif x<5:
        return 'm'
    else:
        return 'l'
```

```python
df['sepal_length2'] = df['petal_length'].apply(function1)
```

```python
sns.scatterplot(data=df, x='sepal_width', y='sepal_length', hue='species', style='sepal_length2')
```

```python
def function2(x):
    if x<3:
        return 3
    elif x<5:
        return 5
    else:
        return 7
```

```python
df['sepal_length3'] = df['sepal_length'].apply(function2)
```

```python
sns.scatterplot(data=df, x='sepal_width', y='sepal_length', hue='species', style='sepal_length3')
```

```python
sns.relplot(data=df, x='sepal_width', y='sepal_length', col='species') # column 별 그래프 작성
```

```python
sns.lmplot(data=df, x='sepal_width', y='sepal_length', hue='species') # 회귀선 작성
```
