```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
pivot_data = df.pivot('month', 'year', 'passengers')
```

```python
pivot_data
```

```python
sns.heatmap(pivot_data)
```

```python
sns.heatmap(pivot_data, linewidths=0.5)
```

```python
sns.heatmap(pivot_data, linewidths=0.5, cmap='Blues')
```

```python
sns.heatmap(pivot_data, linewidths=0.5, cmap='Blues', annot=True) # 각 셀별로 값 표시
```

```python
sns.heatmap(pivot_data, linewidths=0.5, cmap='Blues', annot=True, fmt='d') # 값을 정수형태로 표시
```

```python
fig, (ax, cbar_ax) = plt.subplots(2)
```

```python
ax = sns.heatmap(pivot_data, ax=ax,
                    cbar_ax=cbar_ax,
                    cbar_kws={'orientation':'horizontal'})
```

```python
fig
```

```python
ax = sns.heatmap(pivot_data, ax=ax,
                    cbar_ax=cbar_ax,
                    cbar_kws={'orientation':'vertical'})
```

```python
fig
```

```python
grid_kws = {"height_ratios":(85,1), "hspace":0.4} # heigth_ratios : (a, b) b 는 아래 범례의 굵기, hspace 는 차트와 범례 사이의 간격

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                    cbar_ax=cbar_ax,
                    cbar_kws={'orientation':'horizontal'})
```
