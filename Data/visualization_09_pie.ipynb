```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
mv_title = ['cruella', 'demon slayer', 'path to school']
mv_aud = [664308, 2099131, 20067]

data = {'movie_title' : mv_title, 'movie_audience' : mv_aud}
df = pd.DataFrame(data)
```

```python
plt.pie(df['movie_audience']);
```

```python
plt.pie(df['movie_audience'], labels=df['movie_title']);
```

```python
plt.pie(df['movie_audience'], labels=df['movie_title'], autopct='%0.2f%%');
```

```python
colors_list = ['green', 'blue', 'red']
plt.pie(df['movie_audience'], labels=df['movie_title'], autopct='%0.2f%%', colors=colors_list);
```

```python
explode_list = [0, 0.1, 0]

plt.pie(df['movie_audience'], labels=df['movie_title'], autopct='%0.2f%%', explode=explode_list);
```
