```python
# 한 차트에 멀티 레이블 그리기
# plt.legend 사용법
```

```python
import matplotlib.pyplot as plt
```

```python
x1, y1 = [1,2,3], [1,2,3]
x2, y2 = [1,2,3], [1,4,7]
```

```python
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
```

```python
plt.plot(x1, y1, 'b', x2, y2, 'r')
```

```python
plt.plot(x1, y1, 'b', label='data1')
plt.plot(x2, y2, 'r', label='data2')
plt.legend()
```

```python
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
plt.legend(['data1', 'data2'])
```

```python
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
plt.legend(['data1', 'data2'], loc='upper right')
```

```python
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
plt.legend(['data1', 'data2'], fontsize=20)
```
