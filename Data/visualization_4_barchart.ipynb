```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
mv_title = ['a', 'b', 'c', 'd']
mv_aud = [664308, 2099131, 20067, 0]

data = {'movie_title' : mv_title, 'movie_audience' : mv_aud}
df = pd.DataFrame(data)
```

```python
sns.barplot(data=df, x='movie_title', y='movie_audience')
```

```python
sns.set(rc={'figure.figsize':(10, 5)})
```

```python
sns.barplot(data=df, x='movie_title', y='movie_audience')
```

```python
sns.barplot(data=df, x='movie_title', y='movie_audience',
                order=df.sort_values('movie_audience').movie_title) # 오름차순 정렬
```

```python
sns.barplot(data=df, x='movie_title', y='movie_audience',
                order=df.sort_values('movie_audience', ascending=False).movie_title) # 내림차순 정렬
```

```python
chart = sns.barplot(data=df, x='movie_title', y='movie_audience',
                order=df.sort_values('movie_audience', ascending=False).movie_title)

ylabels = ['{:,.0f}'.format(i) + '0,000' for i in chart.get_yticks() / 10000]
chart.set_yticklabels(ylabels)
```

```python
# 가로로 그리기
```

```python
chart = sns.barplot(data=df, x='movie_audience', y='movie_title',
                    order=df.sort_values('movie_audience', ascending=False).movie_title)

xlabels = ['{:,.0f}'.format(i) + '0,000' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)
```

```python
# 차트 색상 변경
```

```python
chart = sns.barplot(data=df, x='movie_audience', y='movie_title',
                    order=df.sort_values('movie_audience', ascending=False).movie_title,
                    color='blue')

xlabels = ['{:,.0f}'.format(i) + '0,000' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)
```

```python
# 제목 추가, 폰트 사이즈 변경
```

```python
chart = sns.barplot(data=df, x='movie_audience', y='movie_title',
                    order=df.sort_values('movie_audience', ascending=False).movie_title,
                    color='blue')

xlabels = ['{:,.0f}'.format(i) + '0,000' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)

plt.xlabel('movie_audience', fontsize=15)
plt.ylabel('movie_title', fontsize=15)
plt.title('movie')
```
