```python
# linechart 를 이용한 선도표 작성
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.lineplot(data=df, x='year', y='passengers') # 선 : 실제 데이터, 면 : 신뢰구간
```

```python
sns.lineplot(data=df, x='year', y='passengers')
plt.xticks(df['year']); # x 축 부분을 1년 단위로 표시, 세미콜론 입력시 추가 출력 문구 생략 가능
```

```python
sns.lineplot(data=df, x='year', y='passengers')
plt.xticks(df['year'], rotation=45); # x 축 각도를 돌림
```

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month')
```

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month', palette='Set2')
```

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month', palette='Set2', style='month')
```
