```python
# distplot / histplot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
```

```python
sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['sepal_length'], bins=15)
```

```python
sns.histplot(df['sepal_length']) # 갯수
```

```python
sns.histplot(df['sepal_length'], bins=20)
```

```python
sns.histplot(df['sepal_length'], bins=20, kde=True) # 밀도 곡선 표시
```

```python
sns.histplot(df['sepal_length'], bins=20, kde=True, stat='density') # y축 변환
```

```python
sns.histplot(data=df, x='sepal_length', bins=20, kde=True, stat='density', hue='species')
```

```python
sns.histplot(data=df, y='sepal_length', bins=20, kde=True, stat='density', hue='species')
```

```python
sns.histplot(data=df, x='sepal_length', y='sepal_width')
```

```python
sns.histplot(data=df, x='sepal_length', y='sepal_width', cbar=True)
```

```python
sns.displot(df['sepal_length'])
```

```python
sns.displot(data=df, x='sepal_length')
```

```python
sns.displot(data=df, x='sepal_length', height=5, aspect=3) # aspect = width (heigth * aspect)
```

```python
sns.jointplot(data=df, x='sepal_length', y='sepal_width')
```

```python
sns.jointplot(data=df, x='sepal_length', y='sepal_width', kind='hist', cbar=True)
```

```python
sns.jointplot(data=df, x='sepal_length', y='sepal_width', kind='hist', cbar=True, height=10)
```
